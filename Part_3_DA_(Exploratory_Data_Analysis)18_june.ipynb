`2) GroupBy in Python`

- can be applied on categorical variables
- group data in categories
- can group single/multiple variables


# 1st method : GroupBy

```python
#for example, find the average price of vehicles
#             & observe how they differ between diff types of "body-style"
#              and "drive-wheels" variables.

df_test = df[['drive-wheels', 'body-style', 'price']]
df_grp = df_test.groupby(['drive-wheels','body-style'], as_index = False).mean()
df_grp

```

# 2nd method : Pivot( ) method

- to make it easier to understand, we can transform previous table to a pivot table using pivot() method.

```python
# Another method to group is using - Pivot() method

df_pivot = df_grp.pivot(index='drive-wheels', columns='body-style')
df_pivot
```

# 3rd method : Heatmap
- another way to represent pivot table is using a heatmap plot

```python
#change data type 'object' to 'float' first
#heatmap only runs on float dtype
df_pivot = df_pivot.astype(float)
```

```python
#heatmap - can plot target variable over multiple variables
plt.pcolor(df_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
#The default labels convey no useful information to us
#let's change it to meaningful visualization
#row = "body-style"
#column = "drive-wheels"

#for heatmap plots
fig, ax = plt.subplots()
im = ax.pcolor(df_pivot, cmap='RdBu')

#label names
row_labels = df_pivot.columns.levels[1]
col_labels = df_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(df_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(df_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

#show the colored figure of heatmap plots
fig.colorbar(im)
plt.show()
```

`3) Analysis of Variance ANOVA`

- Statistical comparison of groups
- Example: Average price of different vehicle makes

# **ANOVA**

- ANOVA = Analysis of Variance
- used to find correlation between different group of a categorical variable.

- `what can be obtained from ANOVA?`
    - `F-test score` = ratio of variation between group's mean over the variation within each of sample groups
    - `p-value` = show whether the results is statisfically significant

```python
#Example: ANOVA between "Honda" and "Subaru"

import scipy
from scipy import stats
df_anova = df[["make","price"]]
grouped_anova = df_anova.groupby(["make"])
anova_results_1 = stats.f_oneway(grouped_anova.get_group("honda")["price"],
                                 grouped_anova.get_group("subaru")["price"])
anova_results_1
```

From results of Example 1:
- F-test = 0.20
- p-value = 0.66
- hence, The prices between Hondas and Subarus are `not significantly different`
  - bcoz, the F-test
score is less than 1 and p-value is larger than 0.05.

```python
#Example 2: ANOVA between "Honda" and "Jaguar"
anova_results_2 = stats.f_oneway(grouped_anova.get_group("honda")["price"],
                                 grouped_anova.get_group("jaguar")["price"])
anova_results_2
```

From results of Example 2:
- F-test = 401 (which is very large)
- p-value = 1 (which larger than 0.05)
- hence, The prices between Hondas and Jaguar are `significantly different`

In Conclusion:
- there is a strong correlation between a categorical variable and other variables,
- if the ANOVA test gives a results of
`large F-test value` and a `small p-value.`


`4) Correlation`

- measure to what extent different variables are interdependent (correlated)
- For example:
- Lung Cancer >> Smoking
  - high chance of getting lung cancer if you smoke
- Rain >> Umberella
  - if not raining, people wouldn't carry umbrella


```python
#Example 1: Correlation between engine size & car price

# Check and convert data types if necessary
df["engine-size"] = df["engine-size"].astype(float)
df["price"] = df["price"].astype(float)

# Handle missing or invalid values if necessary
df = df.dropna(subset=["engine-size", "price"])

# Create the scatter plot with regression line
sns.regplot(x="engine-size", y="price", data=df)

# Set the lower limit of the y-axis to 0
plt.ylim(0)

# Show the plot
plt.show()
```

`From example 1 plot: `
- the straight line through the data points is very steep, which shows there is a positive linear relationship between the two variables.
  - the `bigger the value of engine size, the higher the price.`
  - Also, `slope of the line = positive`
  - Hence, `there is a postive correlation between "engine-size" and "price".`

```python
#Example 2: Correlation between "highway-mpg" and "price"

# Check and convert data types if necessary
df["highway-mpg"] = df["highway-mpg"].astype(float)

# Handle missing or invalid values if necessary
df = df.dropna(subset=["highway-mpg", "price"])

# Create the scatter plot with regression line
sns.regplot(x="highway-mpg", y="price", data=df)

# Set the lower limit of the y-axis to 0
plt.ylim(0)

# Show the plot
plt.show()
```

`From Example 2 plot: `

  - When the `highway-mpg value is increasing, the value of price goes down.`
  - Hence, `there is a negative linear relationship between highway-mpg and price.`
  - Eventhough the relationship is negative, the slope of line is steep (which means that the highway miles per gallon is still a good predictor of price)
  - These two variables are said to have `a negative correlation.`

```python
#Example 3: Weak Correlation between "peak-rpm" and "price"

# Replace non-numeric values with NaN
df.loc[:, "peak-rpm"] = pd.to_numeric(df["peak-rpm"], errors="coerce")

# Drop rows with NaN values in the "peak-rpm" column
df = df.dropna(subset=["peak-rpm"])

# Convert the "peak-rpm" column to float
df["peak-rpm"] = df["peak-rpm"].astype(float)

# Create the scatter plot with regression line
sns.regplot(x="peak-rpm", y="price", data=df)

# Set the lower limit of the y-axis to 0
plt.ylim(0)

# Show the plot
plt.show()
```

`From Example 3 plot: `
- Both low peak-rpm and high values of peak-rpm have low and high prices,
- therefore, we cannot use peak-rpm to
predict the values.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats

# Assuming you have a DataFrame named 'df' with columns "horsepower" and "price"

# Check and convert data types if necessary
df["horsepower"] = df["horsepower"].astype(float)

# Drop rows with NaN values in the "horsepower" column
df = df.dropna(subset=["horsepower"])

# Calculate Pearson correlation coefficient and p-value
pearson_coef, p_value = stats.pearsonr(df['horsepower'], df['price'])

# Create a correlation matrix
corr_matrix = df.corr()

# Generate the correlation heatmap using Seaborn
sns.heatmap(corr_matrix, annot=False, cmap='RdBu')
#if annot = True, it will showed all the values on the heatmap

# Add a title
plt.title('Correlation Heatmap')

# Show the plot
plt.show()
```

`From this Correlation-Heatmap plot: `

diagonal are `highly correlated (which is 1)`
- This correlation heatmap gives us a good overview of how the different variables are related
