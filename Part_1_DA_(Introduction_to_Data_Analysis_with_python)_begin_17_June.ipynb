- file path of dataset
 - Computer: /Dekstop/mydata.csv
 - Internet: https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data


```python
#import csv into python

import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header = None)
df

#read_csv usually assume data with header, hence we need to specify if the data has no header
#df will show entire dataframe
```

```python
#to print the specific rows that you wish to see:
# 1st, df.head(n) = will show the first (n) rows of dataframe
# 2nd, df.tail(n) = will show the bottom (n) rows of dataframe

df.head(6)

```

```python
df.tail(5)
```

```python
#replace default header using (df.columns = headers)

headers = ["symboling","normalized-loss","make","fuel-type","aspiration",
           "num-of-doors","body-style","drive-wheels","engine-location",
           "wheel-base","length","width","height","curb-weight","engine-type",
           "num-of-cylinders","engine-size","fuel-system","bore","stroke",
           "compression-ratio","horsepower","peak-rpm","city-mpg","highway-mpg",
           "price"
]
df.columns = headers
```

```python
#df.head() function will show the first 5 rows automatically
df.head()
```

```python
#How to export a Pandas dataframe that we have modified to new CSV file
path = "/content/drive/MyDrive/automobile.csv"
#new file path that we want to save

df.to_csv(path)
#this will save the modified dataframe to new csv file
```

- potential info & type mismatch

```python
#to check data types
df.dtypes
```

```python
#returns a statistical summary using describe() method
df.describe()
```

```python
#next, if we put (include = "all") in df.describe() function
#it will provide a full summary statistics (inlcuding for 'object' data type as well)

df.describe(include = "all")
```

`Note:`

- unique = number of rare object in the column
- top = most frequently occuring object
- freq = no. of time the top object appear in the column


- NaN = means not a number

```python
#Another method to check data type is using df.info() fucntion

df.info()
#this will show the datatype for top 30 row
#and bottom 30 row in the column
```
